```python
#Funzione per capire se una stringa rappresenta un intero
def RepresentsInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False
```

```python
#Creazione di una matrice contenente per ogni colonna rispettivamente:  il giorno, mese, anno, ora, minuti e secondi del singolo tweet
import numpy as np
dbtweets['data'] = dbtweets['data'].str.replace(':', ' ')
a = dbtweets['data'].str.split(",").apply(pd.Series)
```

```python
def Hour(hour):
    correct_hour = []
    Checkpoint = "23"
    for i in range(len(hour)):
        str_hour = str(hour[i]).strip()
        if (RepresentsInt(str_hour)==True):
            if ((int(str_hour)) == 22) or ((int(str_hour)) == 23):
                if ((int(str_hour)) == 22):
                    str_hour = "0"
                    correct_hour.append(str_hour)
                    Checkpoint = correct_hour[i]
                else:
                    str_hour = "1"
                    correct_hour.append(str_hour)
                    Checkpoint = correct_hour[i]
            else:
                correct_hour.append(str(int(str_hour)+2))
                Checkpoint = correct_hour[i]
        else:
            correct_hour.append(Checkpoint)
    return correct_hour
```

```python
hour = Hour(a[3])
hour
```

```python
def Day(day, hour):
    correct_day = []
    Checkpoint = "13"
    for i in range(len(hour)):
        str_day = str(day[i]).strip()
        str_hour = str(hour[i]).strip()
        if (RepresentsInt(str_day) == False):
            if ((int(str_hour)) != 0):
                correct_day.append(Checkpoint)
                Checkpoint = correct_day[i]
            else:
                new_day = str(int(Checkpoint)+1)
                correct_day.append(new_day)
                Checkpoint = correct_day[i]
        else:
            correct_day.append(str_day)
            Checkpoint = correct_day[i]
    return correct_day
```

```python
day = Day(a[2], hour)
```

```python
def Minutes(minutes):
    correct_minutes = []
    Checkpoint = "18"
    for i in range(len(minutes)):
        str_minute = str(minutes[i]).strip()
        if (RepresentsInt(str_minute) == False):
                correct_minutes.append(Checkpoint)
                Checkpoint = correct_minutes[i]
        else:
            correct_minutes.append(str_minute)
            Checkpoint = correct_minutes[i]
    return correct_minutes
```

```python
minutes = Minutes(a[4])
```

```python
def Seconds(seconds):
    correct_seconds = []
    Checkpoint = "26"
    for i in range(len(seconds)):
        str_seconds = str(seconds[i]).strip()
        if (RepresentsInt(str_seconds) == False):
            correct_seconds.append(Checkpoint)
            Checkpoint = correct_seconds[i]
        else:
            if(int(str_seconds)<=10 & int(Checkpoint)<=59 & int(Checkpoint)>10):
                split = list(str(Checkpoint))
                str_seconds = split[0] + str_seconds
                correct_seconds.append(str_seconds)
                Checkpoint = correct_seconds[i]
            else:
                correct_seconds.append(str_seconds)
                Checkpoint = correct_seconds[i]
    return correct_seconds
```

### Creazione della libreria 

```python
def Library(Tot_Players):
    players = []
    surname=""
    for i, row in Tot_Players.iterrows():
        surname = Tot_Players['Shirt Name'][i].title()
        tot_name=(Tot_Players['FIFA Popular Name'][i]).split()
        name=tot_name[1]
        name=name[:1]
        players.append(name + ". " + surname)
    return players

Croatia_library = Library(Croatia)

France_library = Library(France)


France_library
```

### Velocity

```python
def Statistics_mean(Statistics):
    for c in Statistics:
        Statistics[c] = pd.to_numeric(Statistics[c], errors='coerce')
    return Statistics.mean(axis=1)
```
